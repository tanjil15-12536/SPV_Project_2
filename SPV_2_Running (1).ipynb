```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
df=pd.read_csv("train.csv")
df
```

```python
df=df.rename(columns = {"Sex" : "Gender"})
df=df.rename(columns = {"Embarked" : "Country"})
df=df.rename(columns = {"SibSp" : "Siblings"})
df=df.rename(columns = {"Fare" : "Ticket Cost"})
df
```

```python
df.dropna(inplace = True)
df
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
df.dtypes
```

```python
df["Country"].replace({"S": "Switzerland", "C": "Canada", "Q": "Qatar"}, inplace=True)
df
```

```python
pd.crosstab(df.Gender, df.Survived)
```

```python
over_30 = df[df["Age"]>30.0]
over_30 
```

```python
df["Pclass"].replace({1: "Business Class", 2: "First Class", 3: "Second Class"}, inplace=True)
df
```

```python
df["Pclass"].replace({"Secon Class":"Second Class"}, inplace=True)
df
```

```python
df["Pclass"].replace(s" : 1,{"Business Clas "First Class" : 2, "Second Class" :3}, inplace=True)
df
```

```python
sample_data = df.loc[1:100]
sample_data
```

```python
sample_data.plot(x="Pclass", y="Ticket Cost", kind="bar",color=["red"])
#plt.xlabel('x') 
plt.ylabel('Ticket Cost') 
plt.title("Ticket Cost by Pclass(Bar Graph)")
sample_data
```

```python

sample_data.plot(x="Survived", y="Age", kind="bar",color=["green"])
plt.title("Survived by Age(Bar Graph)")
sample_data
```

```python
sample_data.plot(x="PassengerId", y="Siblings", kind="scatter",color=["blue"])
plt.title("Sibling by passengerID(Scatter Graph)")
sample_data
```
